```python
import requests

# Define the endpoint URL
url = "http://127.0.0.1:8000/api/resume/parse-resume"

# Define the path to your test resume file
file_path = "Mothkuri_Prince_Resume.pdf"  # Replace with your actual file path

# Open the file in binary mode
with open(file_path, "rb") as file:
    # Define the files dictionary to send in the request
    files = {"file": file}
    
    # Send a POST request to the API
    response = requests.post(url, files=files)

# Print the response from the server
print(response.json())

result = response.json()
```

```python
# pretty print the response
import json
print(json.dumps(response.json(), indent=2))

# print like a json object
PrettyJson = json.dumps(response.json(), indent=2)
```
